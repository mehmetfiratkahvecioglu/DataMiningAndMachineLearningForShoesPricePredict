```python
#Gerekli kütüphaneler
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#Gerekli listeler
url_list = []
prices_list = []
propTitles = []
#Özelliklerin çekilmesi
for i in range(1,60):#sayfa sayısı gelmeli
    url = "https://www.trendyol.com/erkek-ayakkabi-x-g2-c114?pi=" + str(i) #Urlin for döngsü ile güncellenmesi
    r = requests.get(url) #Hazırlanan url'e istek yapılması
    source = BeautifulSoup(r.content,"html.parser") #İstek yapılan sayfanın içeriğinin çekilmesi

    
    urls = source.find_all("div",attrs={"class":"p-card-wrppr with-campaign-view"}) #Tüm verilerin bulunması
    for url in urls:
        url_phone = "https://www.trendyol.com"+url.a.get("href") #Verilerin linklerinin bulunması
        url_list.append(url_phone) #Linklerin kaydedilmesi
        print(url_phone)
        
        r_phone = requests.get(url_phone) #Bulunan linke tekrar istek atılması
        source_phone = BeautifulSoup(r_phone.content,"html.parser") #Veri içeriğinin çekilmesi
        
        properties = source_phone.find_all("li",attrs={"class":"detail-attr-item"}) #Bütün özelliklerin bulunması
        
        for prop in properties:
            prop_title = prop.find_all("span")[0].text
            propTitles.append(prop_title)
    
        
    prices = source.find_all("div",attrs={"class":"prc-box-dscntd"}) #Bütün fiyat özelliklerinin bulunması
    for price in prices:
        prices_list.append(price.text) #Bulunan özellikler arasında gezinme ve listeye yazma
        print(price.text)
    
    
        
    
print(str(len(url_list))+" Adet link bulundu.")
print(str(len(prices_list))+" Adet fiyat bulundu.")
print(str(len(propTitles))+" Adet özellik başlığı bulundu.")
```

```python
#Url ve fiyatları bir data frame yazma
df_urls = pd.DataFrame()
df_urls["urls"] = url_list
df_urls["prices"] = prices_list
```
